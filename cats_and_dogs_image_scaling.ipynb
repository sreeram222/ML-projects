```python
dogset=os.path.join("/content/drive/MyDrive/Colab Notebooks/cats and dogs","dogs_set")
dogset
```

```python
flat_data=[]    #3dimensional to 1 dimensional
target=[]
categories=["cats_set","dogs_set"]

datadir="/content/drive/MyDrive/Colab Notebooks/cats and dogs"
for i in categories:
  print("loading data",i)
  path=os.path.join(datadir,i)

  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))
    img_resized=resize(img_array,(150,150,3))#RGB
    flat_data.append(img_resized.flatten())    #to make i dimensional array
    target.append(categories.index(i))
  print("loaded",i,"successfully")
```

```python
img_array
```

```python
flat_data
```

```python
import pandas as pd
df=pd.DataFrame(flat_data)
#df["Target"]=target
df
```

```python
X=df.values
X
```

```python
import numpy as np
y=np.array(target)
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.svm import SVC
clf=SVC()
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
print(accuracy_score(y_test,y_pred)*100)
cm=ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```
