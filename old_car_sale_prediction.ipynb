```python
dumm=pd.get_dummies(df[['Location',"Fuel_Type","Transmission","Owner_Type"]])
dumm
```

```python
dumm.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dumm["Location_Ahmedabad"]=le.fit_transform(dumm["Location_Ahmedabad"])
dumm["Location_Bangalore"]=le.fit_transform(dumm["Location_Bangalore"])
dumm["Location_Chennai"]=le.fit_transform(dumm["Location_Chennai"])
dumm["Location_Coimbatore"]=le.fit_transform(dumm["Location_Coimbatore"])
dumm["Location_Delhi"]=le.fit_transform(dumm["Location_Delhi"])
dumm["Location_Hyderabad"]=le.fit_transform(dumm["Location_Hyderabad"])
dumm["Location_Jaipur"]=le.fit_transform(dumm["Location_Jaipur"])
dumm["Location_Kochi"]=le.fit_transform(dumm["Location_Kochi"])
dumm["Location_Kolkata"]=le.fit_transform(dumm["Location_Kolkata"])
dumm["Location_Mumbai"]=le.fit_transform(dumm["Location_Mumbai"])
dumm["Fuel_Type_CNG"]=le.fit_transform(dumm["Fuel_Type_CNG"])
dumm["Fuel_Type_Diesel"]=le.fit_transform(dumm["Fuel_Type_Diesel"])
dumm["Fuel_Type_Electric"]=le.fit_transform(dumm["Fuel_Type_Electric"])
dumm["Fuel_Type_LPG"]=le.fit_transform(dumm["Fuel_Type_LPG"])
dumm["Fuel_Type_Petrol"]=le.fit_transform(dumm["Fuel_Type_Petrol"])
dumm["Transmission_Automatic"]=le.fit_transform(dumm["Transmission_Automatic"])
dumm["Transmission_Manual"]=le.fit_transform(dumm["Transmission_Manual"])
dumm["Owner_Type_First"]=le.fit_transform(dumm["Owner_Type_First"])
dumm["Owner_Type_Fourth & Above"]=le.fit_transform(dumm["Owner_Type_Fourth & Above"])
dumm["Owner_Type_Second"]=le.fit_transform(dumm["Owner_Type_Second"])
dumm["Owner_Type_Third"]=le.fit_transform(dumm["Owner_Type_Third"])
dumm["Location_Pune"]=le.fit_transform(dumm["Location_Pune"])
```

```python
for column in dumm.columns:
  dumm[column]=le.fit_transform(dumm[column])
```

```python
df.drop(["Location","Fuel_Type","Transmission","Owner_Type"],inplace=True,axis=1)
```

```python
df1=pd.concat([df,dumm],axis=1)
df1
```

```python
df1.isna().sum()
```

```python
df1["Mileage"]=df1["Mileage"].str.replace("km/kg","")
```

```python
df1["Mileage"]=df1["Mileage"].str.replace("kmpl","")
```

```python
df1["Engine"]=df1["Engine"].str.replace("CC","")
```

```python
df1["Power"]=df1["Power"].str.replace("bhp","")
```

```python
df1["Power"]=df1["Power"].str.replace("null","0")
```

```python
df1["Mileage"]=df1["Mileage"].astype(float)
df1["Engine"]=df1["Engine"].astype(float)
df1["Power"]=df1["Power"].astype(float)
```

```python
df1.isna().sum()
```

```python
df.info()
```

```python
import numpy as np
df1.loc[df1.Power==0,"Power"]=np.nan
df1.loc[df1.Mileage==0,"Mileage"]=np.nan
```

```python
df1.isna().sum()
```

```python
df1["Mileage"].value_counts()#mean
df1["Power"].value_counts()#mean
df1["Engine"].value_counts()#mean
df1["Seats"].value_counts()#mode
```

```python
df1["Mileage"]=df1["Mileage"].fillna(df1["Mileage"].mean())
df1["Power"]=df1["Power"].fillna(df1["Mileage"].mean())
df1["Engine"]=df1["Engine"].fillna(df1["Engine"].mean())
df1["Seats"]=df1["Seats"].fillna(df1["Seats"].mode()[0])
```

```python
y=df1["Price"].values
y
```

```python
x=df1.drop(["Price"],axis=1).values
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
```
