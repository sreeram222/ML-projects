```python
import pandas as pd
df=pd.read_csv("/content/Iris.csv")
df
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Species"]=le.fit_transform(df["Species"])#LabelEncoder is used to classify the data into numericals
df
#Iris-setosa 0
#Iris-veriscolor 1
#Iris-virginica 2
```

```python
X=df.iloc[:,1:5].values
X
#for classification of data
```

```python
y=df.iloc[:,-1].values
y
#classification of data for the prediction
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

sc.fit(X)
X=sc.transform(X)
X

#standard scalar is used to change the data to arrays and to convert it to the range 0-1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()

#GaussianNB contionous values
#MultinomialNB discrete values
#BernoulliNB 0 1

nb.fit(X_train,y_train)
y_pred=nb.predict(X_test)
y_pred
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
#to get the accuracy of the model
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cmm=ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
print(y_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
cmm=ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
svm=SVC(kernel="poly")
svm.fit(X_train,y_train)
```

```python
i_pred=svm.predict(X_test)
print(i_pred)
print(y_test)
```
