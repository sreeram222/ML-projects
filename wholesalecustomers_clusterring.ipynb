```python
import pandas as pd
df=pd.read_csv("/content/Wholesale customers data.csv")
df
```

```python
df.isnull().sum()
```

```python
x=df
x
```

```python
# from sklearn.preprocessing import StandardScaler
# sc=StandardScaler()
# x=sc.fit_transform(x)
# x

from sklearn.preprocessing import normalize
data=normalize(x)
data
```

```python
# wcss=[]
# from sklearn.cluster import KMeans
# for i in range(1,10):
#   k=KMeans(n_clusters=i,random_state=0)
#   k.fit(x)
#   wcss.append(k.inertia_)

x=pd.DataFrame(data,columns=df.columns)
x
```

```python
# from matplotlib import pyplot as plt
# plt.plot(range(1,10),wcss)
# plt.show()

import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
dendrogram=sch.dendrogram(sch.linkage(x,method="average"))
plt.show()
```

```python
# from sklearn.cluster import KMeans
# k=KMeans(n_clusters=6,init="k-means++",random_state=0)
# y_pred=k.fit_predict(x)
# y_pred

from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=3,metric="euclidean",linkage="average")
y_pred=hc.fit_predict(x)
y_pred
```

```python
plt.scatter(x["Milk"],x["Grocery"],c=y_pred)
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(x,y_pred)
score
```
