```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/framingham.csv")
df
```

```python
df.shape
```

```python
df.drop(["education"],axis=1,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
l=["cigsPerDay","BPMeds","heartRate"]
for i in l:
  df[i]=df[i].fillna(df[i].mode()[0])
df.isna().sum()
```

```python
df["glucose"].value_counts()
```

```python
g=["glucose","BMI","totChol"]
for i in g:
  df[i]=df[i].fillna(df[i].mean())
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel="poly")
svc.fit(x_train,y_train)
n_pred=svc.predict(x_test)
n_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
h_pred=dt.predict(x_test)
h_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,h_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
gn=GaussianNB()
gn.fit(x_train,y_train)
j_pred=gn.predict(x_test)
j_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,j_pred)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
from sklearn.svm import SVC
sv=SVC()
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
from sklearn.metrics import confusion_matrix,accuracy_score

l2=[knn,sv,gnb,dt]
for i in l2:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(i)
  print(confusion_matrix(y_test,y_pred))
  print(accuracy_score(y_test,y_pred)*100)
  print(i.predict(sc.transform([[0,42,20,10,34,32,32,12,43,56,43,43,23,12]])))
  print()
```
